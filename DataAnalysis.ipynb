Find odd ad even number in given array

```python
for i in arangeArray:
    if i % 2 == 0:
        print("Even Number___________", i)
    else:
        print("Odd Number____________", i)
```

Vactorization

```python
a = np.arange(0,5)
b = np.arange(5, 10)

res = []
for i in range(len(a)):
    res.append(a[i]+ b[i])

np.array(res)
%timeit np.array(res)
```

```python
a+b
%timeit a+b
```

```python
a/b
%timeit a/b
```

## Pandas

```python
import pandas as pd
```

Create DataFrame  using Json

```python
fruitBowl = {'Fruit':['apple', 'orange', 'grapes', 'banana'], 'Price':['160', '120', '90', '80']}
pd.DataFrame(fruitBowl)
```

Create DataFrame using List

```python
fruitList = ['Apple', 'Orange', 'Grapes', 'Banana', 'Govaba']
priceList = ['160', '120', '90', '80', '125']
dfList = pd.DataFrame(list(zip(fruitList, priceList)))
print(dfList)
```

Creating DataFrame from List of Json Data

```python
listJson=[{'Fruit':'Apple', 'Price':100}, {'Fruit':'Orange', 'Price':200}, {'Fruit':'Grapes', 'Price':'170'}]
dfmultiJson = pd.DataFrame(listJson)
print(dfmultiJson)
```
